```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
from matplotlib import font_manager
from scipy.optimize import curve_fit
from matplotlib.pyplot import cm
import statistics

plt.style.use('seaborn-v0_8-bright')
hfont = {'fontname': 'Times'}
csfont = {'fontname':'Didot'}
font = font_manager.FontProperties(family='Times',
                                   style='normal', size=10)
```

```python
df = pd.read_csv('SF_avgconnect_constants_0301.csv')
p_vals = list(df['p'].drop_duplicates().sort_values())
e_vals = list(df['avg_edges'].drop_duplicates().sort_values())
n_vals = list(df['Num_nodes'].drop_duplicates().sort_values())
```

```python
for p in p_vals:
    plt.figure(figsize=(8, 6), dpi = 150)
    plt.title('Mean Propagation Time for a Scale Free Network with Probability of Infection = {}'.format(p), fontsize = 12, **csfont)
    for e in range(len(e_vals)):
        valid_data_p = df[(df['p'] ==  p) & (df['avg_edges'] == e_vals[e]) & (df['Net_type'] == 'Scale Free')]
        plt.plot(valid_data_p['Num_nodes'], valid_data_p['avg_timesteps'], color = cm.winter(e/len(e_vals)), label = 'Average Degree = {}'.format(e_vals[e]))
        plt.errorbar(valid_data_p['Num_nodes'], valid_data_p['avg_timesteps'], yerr = valid_data_p['std_timesteps'], ls = '', capsize = 5, color = cm.winter(e/len(e_vals)), alpha = 0.5)
    plt.legend(prop = font)
    plt.xlabel('Number of Nodes', fontsize = 12, **hfont)
    plt.ylabel('Mean Propagation Time / Timesteps',fontsize = 12,  **hfont)
    plt.grid()
    plt.yscale('log')
    plt.tight_layout()
    plt.show()
        
```

```python
for e in e_vals:
    plt.figure(figsize=(8, 6), dpi = 150)
    plt.title('Mean ASPL for a Scale Free Network with Average Number of Connections = {}'.format(e), fontsize = 12, **csfont)
    for p in range(len(p_vals)):
        valid_data_p = df[(df['p'] ==  p_vals[p]) & (df['avg_edges'] == e) & (df['Net_type'] == 'Scale Free')]
        plt.plot(valid_data_p['Num_nodes'], 1/valid_data_p['avg_ASPL'], color = cm.winter(p/len(p_vals)), label = 'p = {}'.format(p_vals[p]))
        plt.errorbar(valid_data_p['Num_nodes'], 1/np.array(valid_data_p['avg_ASPL']), yerr = 1/np.array(valid_data_p['std_ASPL']), ls = '', capsize = 5, color = cm.winter(p/len(p_vals)), alpha = 0.5)
    plt.legend(prop = font)
    plt.xlabel('Number of Nodes', fontsize = 12, **hfont)
    plt.ylabel('ASPL',fontsize = 12,  **hfont)
    plt.grid()
    # plt.yscale('log')
    plt.tight_layout()
    plt.show()
        
```

```python
plt.figure(figsize=(8, 6), dpi = 150)
plt.plot(1/df['avg_ASPL'],df['avg_timesteps'], '.', color = 'blue')
plt.errorbar(1/df['avg_ASPL'],df['avg_timesteps'], yerr = df['std_timesteps'], ls = '', capsize = 4, alpha = 0.2, color = 'blue')
plt.xlabel('ASPL', fontsize = 12, **hfont)
plt.ylabel('Mean Propagation Time / Timesteps', fontsize = 12, **hfont)
plt.tight_layout()
plt.grid()
plt.show()
```
