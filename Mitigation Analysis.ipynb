## Mit Proportion vs prob of outbreak

```python
dfs = [df_70, df_75, df_82, df_85, df_87, df_90, df_92, df_95, df_97]
#aspldfs = [df_75, df_70]
labels = [r"Vaccine Effectiveness = 70%",r"Vaccine Effectiveness = 75%", r"Vaccine Effectiveness = 82%",
         r"Vaccine Effectiveness = 85%", r"Vaccine Effectiveness = 87%",r"Vaccine Effectiveness = 90%", 
          r"Vaccine Effectiveness = 92%",r"Vaccine Effectiveness = 95%",r"Vaccine Effectiveness = 97%"]


def sigmoid(x ,x0, k, b):
    y = -1 / (1 + np.exp(-k*(x-x0))) + b
    return (y)

for i in range(len(dfs)):
    plt.scatter(dfs[i]["mit_prop"], dfs[i]["proboutbreak"], 
                 color = cm.viridis(i/len(dfs)), label = '{}'.format(labels[i]), ls= 'dashed', alpha = 0.3)
    popt1, pcov1 = curve_fit(sigmoid, dfs[i]["mit_prop"], dfs[i]["proboutbreak"])
    print(popt1)
    print(pcov1)

    xx = np.linspace(0, 1)
    plt.plot(xx, sigmoid(xx, *popt1), color = cm.viridis(i/3), ls = '--')
    plt.fill_between(xx, sigmoid(xx, *[popt1[i] if i != len(popt1)-1 else popt1[i] + np.sqrt(pcov1[i][i]) for i in range(len(popt1))]), 
                        sigmoid(xx, *[popt1[i] if i != len(popt1)-1 else popt1[i] - np.sqrt(pcov1[i][i]) for i in range(len(popt1))]), color = cm.viridis(i/3), alpha = 0.2)

                 
plt.xlabel('Proportion of Population Mitigating', fontsize = 12, **hfont)
plt.ylabel('Probability of Outbreak',fontsize = 12,  **hfont)
plt.title('Probability of Outbreak against Proportion \n of Population Mitigating', fontsize = 16, **csfont)
plt.legend(prop = font)
plt.grid()
#plt.ylim(0,3)
#plt.yscale('log')
plt.tight_layout()
plt.show()
```
