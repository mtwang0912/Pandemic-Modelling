```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

Node_number = [100,1000,10000,50000,100000,200000,300000]

def mean(array):
    mean_array = [] 
    for i in range(len(array)):     
        mean_array.append(np.mean(array[i]))            
    return mean_array

def std(array):
    std_array = []
    for i in range(len(array)):     
        std_array.append(np.std(array[i]))            
    return std_array

# Control Variables: 

# family_weight = 0.5
# stranger_weight = 0.2
# x_value = 6  
# prob_of_infection = 0.5
#initial sim stopped after 5 timesteps

#Independent variable: 1 node decreased per edge


#weighted

N_100_w_frac = [0.02, 0.06, 0.09, 0.02, 0.03]
N_1000_w_frac = [0.006, 0.009, 0.019, 0.004, 0.006]
N_10000_w_frac = [0.0269, 0.0009, 0.0006, 0.0002, 0.0003]
N_50000_w_frac = [2e-05, 0.00014, 0.00028, 4e-05, 0.00014]
N_100000_w_frac = [0.00032, 4e-05, 4e-05, 0.00014, 0.0001]
N_200000_w_frac = [0.001045, 1.5e-05, 0.000245, 5e-06, 1e-05]
N_300000_w_frac = [1e-05, 8e-05, 0.00161, 0.00027666666666666665, 2e-05]
                   

N_100_w_timestep = [43, 42, 34, 46, 38]
N_1000_w_timestep = [58, 54, 63, 59, 59]
N_10000_w_timestep = [95, 71, 87, 80, 90]
N_50000_w_timestep = [112, 99, 94, 94, 112]
N_100000_w_timestep = [107, 96, 101, 110, 104]
N_200000_w_timestep = [93, 119, 111, 127, 101]
N_300000_w_timestep = [99, 103, 100, 113, 111]
                   
#unweighted
                   
N_100_unw_frac = [0.19, 0.26, 0.03, 0.31, 0.15]
N_1000_unw_frac = [0.332, 0.149, 0.009, 0.002, 0.024]
N_10000_unw_frac = [0.0001, 0.0005, 0.001, 0.083, 0.001]
N_50000_unw_frac = [0.01684, 0.00026, 0.00036, 0.00204, 0.00028]
N_100000_unw_frac = [0.00028, 0.00807, 0.00035, 0.00522, 0.00051]
N_200000_unw_frac = [0.00216, 0.007675, 0.043585, 0.00315, 0.000175]
N_300000_unw_frac = [0.00011666666666666667, 7.333333333333333e-05, 9.333333333333333e-05, 0.00030333333333333335, 0.00037]


N_100_unw_timestep = [19, 25, 38, 24, 21]
N_1000_unw_timestep = [31, 27, 31, 54, 29]
N_10000_unw_timestep = [63, 42, 51, 44, 48]
N_50000_unw_timestep = [53, 56, 65, 57, 66]
N_100000_unw_timestep = [55, 55, 56, 52, 62]
N_200000_unw_timestep = [52, 59, 57, 62, 58]
N_300000_unw_timestep = [58, 61, 73, 60, 61]



unw_frac = [N_100_unw_frac,N_1000_unw_frac,N_10000_unw_frac,N_50000_unw_frac,
            N_100000_unw_frac,N_200000_unw_frac,N_300000_unw_frac]
                     
unw_timestep = [N_100_unw_timestep, N_1000_unw_timestep, N_10000_unw_timestep, N_50000_unw_timestep,
 N_100000_unw_timestep, N_200000_unw_timestep, N_300000_unw_timestep]

w_frac = [N_100_w_frac, N_1000_w_frac, N_10000_w_frac, N_50000_w_frac,
 N_100000_w_frac, N_200000_w_frac, N_300000_w_frac]

w_timestep = [N_100_w_timestep, N_1000_w_timestep, N_10000_w_timestep, N_50000_w_timestep,
 N_100000_w_timestep, N_200000_w_timestep, N_300000_w_timestep]


def plot_frac(unweighted, unweighted_error, weighted, weighted_error):
    fig = plt.figure(figsize=(3, 2), dpi=500)
    ax = fig.add_subplot(111)
    ax.grid()

    ax.errorbar(unweighted, Node_number, xerr=unweighted_error, fmt='.', label='Unweighted', color='blue',
                capsize=1, ms=1, elinewidth=1, capthick=1)
    ax.errorbar(weighted, Node_number, xerr=weighted_error, fmt='.', label='Weighted', color='green',
                capsize=1, ms=1, elinewidth=1, capthick=1)
    ax.set_ylim(0, 400000)
    ax.set_xlim(-0.05,0.05)
    ax.set_xlabel('Fraction', fontsize=8)
    ax.set_ylabel('Total Number of Nodes', fontsize=8)

    # Change the title based on the family_weights array
    
    
    # You can choose any method to determine the value
    ax.set_title(f' fraction of infected/uninfected for max timestep = 5', fontsize=10)

    ax.plot(unweighted, Node_number, "--", color='blue', alpha=0.5, linewidth=0.7)
    ax.plot(weighted, Node_number, "--", color='green', alpha=0.5, linewidth=0.7)
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles[::-1], labels[::-1], loc='upper left', fontsize=6, framealpha=1)
    ax.tick_params(axis='both', labelsize=6, direction='in', top=True, right=True, which='both')

    ax.xaxis.set_minor_locator(MultipleLocator(2))
    ax.yaxis.set_minor_locator(MultipleLocator(10000))
    
    return fig.tight_layout()

def plot_2(unweighted, unweighted_error, weighted, weighted_error):
    fig = plt.figure(figsize=(3, 2), dpi=500)
    ax = fig.add_subplot(111)
    ax.grid()

    ax.errorbar(unweighted, Node_number, xerr=unweighted_error, fmt='.', label='Unweighted', color='blue',
                capsize=1, ms=1, elinewidth=1, capthick=1)
    ax.errorbar(weighted, Node_number, xerr=weighted_error, fmt='.', label='Weighted', color='green',
                capsize=1, ms=1, elinewidth=1, capthick=1)
    ax.set_ylim(0, 400000)
    ax.set_xlabel('Timestep', fontsize=8)
    ax.set_ylabel('Total Number of Nodes', fontsize=8)
    
   
    ax.set_title('Final timestep after decreasing edges of nodes by 1' , fontsize=10)

    ax.plot(unweighted, Node_number, "--", color='blue', alpha=0.5, linewidth=0.7)
    ax.plot(weighted, Node_number, "--", color='green', alpha=0.5, linewidth=0.7)
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles[::-1], labels[::-1], loc='upper left', fontsize=6, framealpha=1)
    ax.tick_params(axis='both', labelsize=6, direction='in', top=True, right=True, which='both')

    ax.xaxis.set_minor_locator(MultipleLocator(2))
    ax.yaxis.set_minor_locator(MultipleLocator(10000))

    return fig.tight_layout()

plot_frac(mean(unw_frac), std(unw_frac), mean(w_frac), std(w_frac))
plot_2(mean(unw_timestep), std(unw_timestep), mean(w_timestep), std(w_timestep))

                   
```
